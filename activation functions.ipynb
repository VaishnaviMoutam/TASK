```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(-10, 10, 100)

# Sigmoid activation and derivative
sigmoid = 1 / (1 + np.exp(-x))
sigmoid_derivative = sigmoid * (1 - sigmoid)

# ReLU activation and derivative
relu = np.maximum(0, x)
relu_derivative = np.where(x <= 0, 0, 1)

# Tanh activation and derivative
tanh = np.tanh(x)
tanh_derivative = 1 - tanh**2

# softmax activation and derivative
softmax_x = np.exp(x) / np.sum(np.exp(x))
derivative = softmax_x * (1 - softmax_x)




plt.subplot(2, 2, 1)
plt.plot(x, sigmoid, label='Sigmoid')
plt.plot(x, sigmoid_derivative, label='Derivative', linestyle='--')
plt.legend() 

plt.subplot(2, 2, 2)
plt.plot(x, relu, label='ReLU')
plt.plot(x, relu_derivative, label='Derivative', linestyle='--')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x, tanh, label='Tanh')
plt.plot(x, tanh_derivative, label='Derivative', linestyle='--')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(x, softmax, label='Softmax')
plt.plot(x, softmax_derivative, label='Derivative', linestyle='--')
plt.legend()

plt.tight_layout()
plt.show()


```
