```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Data points
data = np.array([-6 ,-5, -3, -2 ,2,3,5,6])

# Mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

# Calculate Z-scores
Z_scores = (data - mean) / std_dev

# Plotting the standard normal distribution
x = np.linspace(-5, 5, 100)  # Generate values for the x-axis
y = norm.pdf(x, 0, 1)        # PDF values for the standard normal distribution

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standard Normal Distribution')

# Plot the Z-scores as vertical lines
for Z in Z_scores:
    plt.axvline(Z, color='r', linestyle='--')

plt.xlabel('Z-score')
plt.ylabel('Probability Density Function (PDF)')
plt.title('Standard Normal Distribution')
plt.legend()
plt.show()
print(Z_scores)
print(mean)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Load data from CSV
file_path = '/home/vaishnavi-moutam/Downloads/archive(4)/CompanyABCProfit.csv'
df = pd.read_csv(file_path)

# Assuming the column of interest is 'Profit'
data = df['Profit(Rs \'000)'].dropna()  # Drop any NaN values if present

# Calculate mean and standard deviation of the original data
mean = np.mean(data)
std_dev = np.std(data)

# Calculate Z-scores (standardized values)
Z_scores = (data - mean) / std_dev

# Calculate the mean and standard deviation of the Z-scores
standardized_mean = np.mean(Z_scores)
standardized_std_dev = np.std(Z_scores)

# Plotting the standard normal distribution
x = np.linspace(-4, 4, 1000)  # Generate values for the x-axis
y = norm.pdf(x, 0, 1)         # PDF values for the standard normal distribution

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standard Normal Distribution')

# Plot the Z-scores as vertical lines
for Z in Z_scores:
    plt.axvline(Z, color='r', linestyle='--')

plt.xlabel('Z-score')
plt.ylabel('Probability Density Function (PDF)')
plt.title('Standard Normal Distribution with Z-scores')
plt.legend()
plt.grid(True)
plt.show()

# Print the mean and standard deviation of the original data
print(f"Mean of the original data: {mean:.2f}")
print(f"Standard Deviation of the original data: {std_dev:.2f}")

# Print the mean and standard deviation of the standardized data (Z-scores)
print(f"Mean of the standardized data: {standardized_mean:.2f}")
print(f"Standard Deviation of the standardized data: {standardized_std_dev:.2f}")
```
